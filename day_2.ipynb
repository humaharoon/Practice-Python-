### 3. Comparison (Relational Operators )
relationa operator compares two values and return a boolean(true or false).

== equal to

!= not equal to

 (>) greater than

(<) less than

(>=) greater than or equal to

(<=) less than or equal to

#### forexample

```python
a = 11
b = 3 

print( a == b )
print( a != b )
print( a > b )
print(a < b)
print( a >= b )
print(a <= b)
```

### logical operator
logical operators combine conditional statements and return true or false.
#### and
true if both conditions are true.
#### or 
true if atleast one condition is true
#### not
 reverse the boolean value
 ##### forexample

```python
4. # Logical Operators and, or, not
# and -> multiplication -> 0 - > false ,1 -> true 
# or -> addition 
# not -> negation

p = 10
q = 4 

print ( (p > 5) and (q < 3) )  # True and false  -> false
print ( (p > 5) or (q < 3) )   # True or false  -> true
print ( not(p > 5) )            # not true -> false
```

### 5. Membership operator
membership operator checks if a value exist in a sequence and return true or false.

#### in opertator
return true if value is present
### not in operator
return true if value is not present

```python
# Membership Operators

fruits = [ "banana", "cherry"]
```

```python
print("apple"  in fruits)  # in operator--> false
```

```python
print("apple" not in fruits) #not in operator--> true
```
